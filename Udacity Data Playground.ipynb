```python
import pandas as pd
from IPython.display import display, HTML

df = pd.read_csv('udacity_data/driving_log-copy.csv')
display(df)
```

```python
from sklearn.model_selection import train_test_split
df_train,df_valid = train_test_split(df, test_size=0.2)

print('Length of df_train:', len(df_train))
print('Length of df_valid:', len(df_valid))
```

```python
import numpy as np
df_valid.index
x= list(df_valid.index)
display(df_valid)
```

```python
df_samples = df_valid
print(len(df_samples))
```

```python
# How many steering angles equal 0?
zero_idx = df_samples[df_samples['steering'] == 0].index
print(len(zero_idx))

# Grab subset of zero steer data
pkeep = 0.1
drop_idx = np.random.choice(zero_idx, int( len(zero_idx) * (1-pkeep) ))
print(len(drop_idx))
```

```python
df_final = df_samples.drop(drop_idx)
print(len(df_final))
display(df_final)
```
